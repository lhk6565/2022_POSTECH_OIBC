```python
def to_weather2(data, id):
    data = data[data['id']==id]
    data = to_local(data)
    data = data.dropna()
    data['time'] = pd.to_datetime(data['time'], format='%Y-%m-%d %H:%M:%S')
    data['date'] = data['time'].dt.date
    data['mon'] = data['time'].dt.month
    data['day'] = data['time'].dt.day
    data['hour'] = data['time'].dt.hour
    data['min'] = data['time'].dt.minute
    for i in range(0,len(data)):
        if (data['min'].iloc[i]> 0):
            data['hour'].iloc[i] += 1
            if(data['hour'].iloc[i] == 24):
                data['hour'].iloc[i] = 0
                data['day'].iloc[i] += 1
        data['time'] = data['date'].astype(str) + " " + data['hour'].astype(str) + ':00:00'
    data['time'] = pd.to_datetime(data['time'], format='%Y-%m-%d %H:%M:%S')
    data = data.groupby(['time'],as_index=False).mean()
    return data
```
